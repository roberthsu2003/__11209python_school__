```python
class Person():
    def __init__(self,name,age=18,weight=70,height=170):
        self.name = name
        self.age = age
        self.weight=weight
        self.height=height

    def description(self):
        print(f"姓名:{self.name}")
        print(f"age:{self.age}")
        print(f"體重:{self.weight}")
        print(f"身高:{self.height}")

p1 = Person(name="徐國堂")
p1.description()
```
