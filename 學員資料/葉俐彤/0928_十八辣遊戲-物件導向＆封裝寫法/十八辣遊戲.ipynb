```python
#import一定要寫在最上面，只要執行一次就好
import random
import pyinputplus as pyip
class Player: #設定屬性，四個骰子的內容不可再寫入
    def __init__(self, name:str) :
        self.name = name
        self.__dice1 = 0
        self.__dice2 = 0
        self.__dice3 = 0
        self.__dice4 = 0

#self的意思：誰去執行它？
#如果是p1.roll_dice = 就是p1執行它
#method就是function，def可以寫好幾個，重點是要簡化你的程式
#所以下面def可以改善的地方：def可以分開(骰骰子、判斷各一個)、判斷式內可以先把四個骰子變成list可以簡化內容

#method
    def __play(self) -> int: #骰骰子＋運算結果
        while (True):
            self.__dice1 = random.randint(1, 6) 
            self.__dice2 = random.randint(1, 6)
            self.__dice3 = random.randint(1, 6)
            self.__dice4 = random.randint(1, 6)
            value = 0
            if self.__dice1 == self.__dice2 == self.__dice3 == self.__dice4: #四個數字相同
                value = {4:13, 8:14, 12:15, 20:17, 24:18}
                break
            elif self.__dice1 == self.__dice2 == self.__dice3 or self.__dice1 == self.__dice2 == self.__dice4 or self.__dice1 == self.__dice3 == self.__dice4 or self.__dice2 == self.__dice3 == self.__dice4: #三個數字相同，要重新骰
                continue
            elif self.__dice1 == self.__dice2 and self.__dice3 == self.__dice4 and self.__dice1 != self.__dice3: #兩對
                if self.__dice1 > self.__dice3:
                    value = self.__dice1 + self.__dice2
                    break
                else:
                    value = self.__dice3 + self.__dice4
                    break
            elif self.__dice1 == self.__dice3 and self.__dice2 == self.__dice4 and self.__dice1 != self.__dice2:
                if self.__dice1 > self.__dice2:
                    value = self.__dice1 * 2
                    break
                else:
                    value = self.__dice2 * 2 
                    break
            elif self.__dice1 == self.__dice2 and self.__dice3 != self.__dice4:
                value = self.__dice3 + self.__dice4
                break
            elif self.__dice1 == self.__dice3 and self.__dice2 != self.__dice4:
                value = self.__dice2 + self.__dice4
                break
            elif self.__dice1 == self.__dice4 and self.__dice2 != self.__dice3:
                value = self.__dice2 + self.__dice3
                break
            elif self.__dice2 == self.__dice3 and self.__dice1 != self.__dice4:
                value = self.__dice1 + self.__dice4
                break
            elif self.__dice2 == self.__dice4 and self.__dice1 != self.__dice3:
                value = self.__dice1 + self.__dice3
                break
            elif self.__dice3 == self.__dice4 and self.__dice1 != self.__dice2:
                value = self.__dice1 + self.__dice2
                break
            elif self.__dice1 != self.__dice2 != self.__dice3 != self.__dice4: #四個不相同
                continue
        descript = ""
        descript += f"骰子1 = {self.__dice1}\n"
        descript += f"骰子2 = {self.__dice2}\n"
        descript += f"骰子3 = {self.__dice3}\n"
        descript += f"骰子4 = {self.__dice4}\n"
        descript += f"得分  = {value}"
        return descript
    
    @property #onlyread結果 #呼叫__play()結果
    def value(self) -> int:
        return self.__play
    
    @property
    def __repr__(self) -> str: #如果確定可以計分，計算&傳出結果
        name = f"{self.name}\n"
        return name


# 如果有兩個玩家可以分勝負
if __name__ == "__main__":
    while (True):
        p1 = Player("robert")
        print(p1.__repr__)
        print(p1.value(),"\n")

        p2 = Player("Rachel")
        print(p2.__repr__)
        print(p2.value(), "\n")

        ask = pyip.inputYesNo(prompt="是否要再玩一次(y, n)?\n", yesVal="y", noVal="n")
        if ask == "n":
            print("遊戲結束，請結算香腸！\n")
            break
           
            
```

```python
#其他寫法
import random
import statistics

class Player:
    def __init__(self, name):
        self.__dice1 = 0
        self.__dice2 = 0
        self.__dice3 = 0
        self.__dice4 = 0
        self.name = name

    def __play(self) -> int:
        Dice = [self.__dice1, self.__dice2, self.__dice3, self.__dice4]

        if len(set(Dice)) == 1:
            leopard={4:13,8:14,12:15,16:16,20:17,24:18}
            Dice_sum=sum(Dice)
            score=leopard.get(Dice_sum)
        elif len(set(Dice)) == 4 or Dice.count(self.__dice1) == 3 or Dice.count(self.__dice2) == 3 or Dice.count(
                self.__dice3) == 3 or Dice.count(self.__dice4) == 3:
            score = 0
        else:
            score = sum(Dice) - min(statistics.multimode(Dice)) * 2

        return score

    @property
    def value(self) -> int:
        return self.__play()
    
    #不用return也可以傳出值＞因為把值儲存在self.__dice
    def roll_dice(self):
        self.__dice1 = random.randint(1, 6)
        self.__dice2 = random.randint(1, 6)
        self.__dice3 = random.randint(1, 6)
        self.__dice4 = random.randint(1, 6)

    def __repr__(self) -> str:
        descript = ""
        descript += f"{self.name}\n"
        descript += f"骰子1={self.__dice1}\n"
        descript += f"骰子2={self.__dice2}\n"
        descript += f"骰子3={self.__dice3}\n"
        descript += f"骰子4={self.__dice4}\n"
        descript += f"得分={self.value}分"
        return descript

def main():
    p1 = Player("Robert")
    p2 = Player("John")

    while True:
        p1.roll_dice()
        p2.roll_dice()

        print(p1)
        print(p2)

        user_input = input("按 Enter 繼續遊戲，按 'q' 键退出遊戲: ")
        if user_input == 'q':
            break

    print("遊戲结束")

if __name__ == "__main__":
    main()
```

```python
#其他寫法
import random
import pyinputplus as pyip

class Player:
    def __init__(self, name: str):
        self.name = name
        self.__dice = [0, 0, 0, 0]

    def roll_dice(self):
        self.__dice = [random.randint(1, 6) for _ in range(4)] 
        return self.__dice[0], self.__dice[1], self.__dice[2], self.__dice[3]

    def calculate_score(self):
        value = 0
        while True:
            if all(x == self.__dice[0] for x in self.__dice):
                value = self.__dice[0] * 4
                return value
            elif any(self.__dice.count(x) == 3 for x in self.__dice):
                self.roll_dice()
                continue
            elif self.__dice.count(self.__dice[0]) == 2 and self.__dice.count(self.__dice[2]) == 2:
                value = sum(self.__dice) - (min(self.__dice) * 2)
                return value
            elif self.__dice.count(self.__dice[0]) == 2 and self.__dice.count(self.__dice[1]) == 2:
                value = sum(self.__dice) - (min(self.__dice) * 2)
                return value
            elif self.__dice.count(self.__dice[0]) == 2 and self.__dice.count(self.__dice[3]) == 2:
                value = sum(self.__dice) - (min(self.__dice) * 2)
                return value
            elif any(self.__dice.count(x) == 2 for x in self.__dice):
                value = sum(x for x in self.__dice if self.__dice.count(x) == 1)
                return value
            elif any(self.__dice.count(x) == 1 for x in self.__dice):
                self.roll_dice()
                continue

    def __repr__(self):
        return f"{self.name}: {self.__dice}, Score: {self.calculate_score()}"

if __name__ == "__main__":
    while True:
        p1 = Player("Robert")
        p1.roll_dice()
        print(p1.__repr__())

        p2 = Player("Rachel")
        p2.roll_dice()
        print(p2.__repr__())

        ask = pyip.inputYesNo(prompt="是否要再玩一次(y, n)?\n", yesVal="y", noVal="n")
        if ask == "n":
            print("遊戲結束，請結算香腸！\n")
            break
```
