```python
# 十八啦遊戲
# 四個骰子隨機骰出四個數字
# 當有兩個骰子數字相同時，另外兩個骰子數字相加為分數
# 當有三個骰子數字相同時，不算分數
# 當有四個骰子數字相同時，叫做豹子(6666>5555>4444>3333>2222>1111)
# 四個骰子不一樣的時候，不算分數
# 四個骰子的點數分成兩組成對時，以數字高者分數為主計算
# 不算分數的要重新骰
```

```python
# 可以隨機骰出四個數字的骰子工具
# 將結果分成幾類：
    # 四個一樣：豹子
    # 三個一樣：不算分數
    # 兩個一樣&兩個一樣：大者相加得分
    # 兩個一樣&兩個不一樣：另外兩個不相同的數字相加得分
    # 四個不一樣：不算分數
# 計算成績&重新骰


import random
while (True):
    a = random.randint(1, 6)
    b = random.randint(1, 6)
    c = random.randint(1, 6)
    d = random.randint(1, 6)
    print(f"十八啦！您的成績是：{a, b, c, d}")

    while (True):
        if a == b == c == d: #四個相同
            print(f"恭喜你！你骰出的是「豹子」")
            break
        elif a == b == c or a == c == d or b == c == d: #三個數字相同
            print(f"你得到零分")
            break
        elif a == b and c == d and a != c: #兩個相同&兩者相同
            if a > c:
                print(f"恭喜你！你的成績是{a + b}")
                break
            else:
                print(f"恭喜你！你的成績是{c + d}")
                break
        elif a == c and b == d and a != b:
            if a > b:
                print(f"恭喜你！你的成績是{a * 2}")
                break
            else:
                print(f"恭喜你！你的成績是{b * 2}")
                break
        elif a == d and b == c and a != b:
            if a > b:
                print(f"恭喜你！你的成績是{a * 2}")
                break
            else:
                print(f"恭喜你！你的成績是{b * 2}")
                break
        elif a == b and c != d:
            print(f"恭喜你！你的成績是{c + d}")
            break
        elif a == c and b != d:
            print(f"恭喜你！你的成績是{b + d}")
            break
        elif a == d and b != c:
            print(f"恭喜你！你的成績是{b + c}")
            break
        elif b == c and a != d:
            print(f"恭喜你！你的成績是{a + d}")
            break
        elif b == d and a != c:
            print(f"恭喜你！你的成績是{b + d}")
            break
        elif c == d and a != b:
            print(f"恭喜你！你的成績是{a + b}")
            break
        elif a != b != c != d: #四個不相同
            print("你得到零分")
            break

    a = input("是否要再玩一次？(y,n)")
    if not a == 'y':
        break
```
